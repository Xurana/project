```python
t = np.linspace(0, 10, 51)
print(t)
```

```python
f = np.cos(t)
print(f)
```

```python
plt.axis([0.5, 9.5, -2.5, 2.5])
plt.plot(t, f, color = 'green')
plt.title('График f(t)', fontsize=16, color='DarkRed')
plt.xlabel('Значения t')
plt.ylabel('Значения f')
plt.show()
```

```python
x = np.linspace(-3, 3, 51)
print(x)
```

```python
title_dict = {'fontsize': '14', 'color' : 'Blue', 'family': 'Calibri'}
y1 = x**2
y2 = 2 * x + 0.5
y3 = -3 * x - 1.5
y4 = np.sin(x)

fig, ax = plt.subplots(nrows=2, ncols=2)
ax1, ax2, ax3, ax4 = ax.flatten()

ax1.plot(x, y1)
ax1.set_xlim(-5, 5)
ax1.set_title('График y1', fontdict = title_dict)

ax2.plot(x, y2)
ax2.set_title('График y2', fontdict = title_dict)

ax3.plot(x, y3)
ax3.set_title('График y3', fontdict = title_dict)

ax4.plot(x, y4)
ax4.set_title('График y4', fontdict = title_dict)

fig.set_size_inches(8, 6)
plt.subplots_adjust(wspace = 0.3, hspace= 0.3)
```

```python
import pandas as pd
plt.style.use('fivethirtyeight')
```

```python
DATASET_PATH = 'creditcard.csv'
```

```python
df = pd.read_csv(DATASET_PATH, sep = ',')
df.head()
```

```python
classes_list = df['Class'].value_counts()
print(classes_list)
```

```python
classes_list.plot(kind='barh')
```

```python
classes_list.plot(kind='barh', logx = True)
```

```python
class0 = df.loc[df['Class'] == 0, ['V1']]
class1 = df.loc[df['Class'] == 1, ['V1']]
```

```python
plt.hist(class0['V1'], bins=20, density=True, alpha=0.5, label='Class 0', color='grey')
plt.hist(class1['V1'], bins=20, density=True, alpha=0.5, label='Class 1', color='red')
plt.legend()
plt.show()
```
